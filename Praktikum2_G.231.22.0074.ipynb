```python
def selectionSort(alist):
 for slot in range(0, len(alist) - 1):
        posisi = slot
        for lokasi in range(slot + 1, len(alist)):
            if alist[lokasi] < alist[posisi]:
                posisi = lokasi
        temp = alist[slot]
        alist[slot] = alist[posisi]
        alist[posisi] = temp

angka = [44, 11, 17, 50, 25, 22, 4]
selectionSort(angka)
print(angka)
```

**tupel**

```python
# Membuat tupel dengan daftar klub favorit
klub_favorit = ("Juventus", "Real Madrid", "Liverpool")

# Mengakses elemen tupel
print("klub favorit pertama:", klub_favorit[0])
print("klub favorit kedua:", klub_favorit[1])
print("klub favorit ketiga:", klub_favorit[2])

# Menampilkan semua klub favorit
for klub in klub_favorit:
    print(klub)

# Mengetahui panjang tupel
print("Jumlah klub favorit:", len(klub_favorit))
```

**Dictionary**

```python
# Membuat data
data = {
    "nama": "Carl Johnson ",
    "umur": 35,
    "alamat": "Jl. Grove street no 5",
    "cita-cita": "TNI"
}

# Mengakses nilai
print("Nama:", data["nama"])
print("Umur:", data["umur"])
print("Alamat:", data["alamat"])
print("Cita-cita:", data["cita-cita"])

# Mengubah nilai
data["umur"] = 40
data["alamat"] = "Jl. Doherty no 5"

# Menambahkan entri baru
data["pekerjaan"] = "Ketua Geng"
data["email"] = "carljohnson56@gmail.com"

# Menghapus entri
del data["cita-cita"]

# Menampilkan seluruh isi data
print("\nData Terbaru :")
for kunci, nilai in data.items():
    print(kunci + ": " + str(nilai))
```
